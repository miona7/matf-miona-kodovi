```python
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
data
```

```python
data.isna().any()
```

```python
data['Unnamed: 32'].isna().all()
```

```python
data.drop('Unnamed: 32', axis = 1, inplace = True)
data.drop('id', axis = 1, inplace = True)
data
```

```python
data.columns
```

```python
# y = pd.DataFrame(data['diagnosis'])
y = data['diagnosis']
X = data.drop('diagnosis', axis = 1)
feature_names = X.columns
```

```python
# y.loc[y['diagnosis'] == 'B', 'diagnosis'] = 0
# y.loc[y['diagnosis'] == 'M', 'diagnosis'] = 1
# y
y.replace('B', 0, inplace = True)
y.replace('M', 1, inplace = True)
```

```python
X.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, stratify = y, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from matplotlib import pyplot as plt
```

```python
dtc = DecisionTreeClassifier()
```

```python
# nece radi kad je y dataframe
dtc.fit(X_train, y_train)
```

```python
plt.figure(figsize = (40, 40))
_ = plot_tree(dtc, class_names = ['B', 'M'], feature_names = feature_names, filled = True)
```

```python
plt.figure(figsize = (20, 10))
pd.Series(dtc.feature_importances_, index = feature_names).plot.barh()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
pd.DataFrame(confusion_matrix(y_train, dtc.predict(X_train)), index = ['B', 'M'], columns = ['B', 'M'])
```

```python
pd.DataFrame(confusion_matrix(y_test, dtc.predict(X_test)), index = ['B', 'M'], columns = ['B', 'M'])
```
