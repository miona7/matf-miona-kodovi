```python
import pandas as pd
```

```python
data = pd.read_csv('balloons.csv')
data
```

```python
X = data.drop('inflated', axis = 1)
feature_names = X.columns
y = data['inflated']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, stratify = y, random_state = 43)
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()
oe.fit(X_train)
```

```python
X_train = pd.DataFrame(oe.transform(X_train), columns = feature_names)
X_test = pd.DataFrame(oe.transform(X_test), columns = feature_names)
```

```python
from sklearn.naive_bayes import CategoricalNB
```

```python
cnb = CategoricalNB()
```

```python
cnb.fit(X_train, y_train)
```

```python
cnb.category_count_   # p(Xi|y)
```

```python
cnb.class_count_      # p(y)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
pd.DataFrame(confusion_matrix(y_train, cnb.predict(X_train)), columns = ['T', 'F'], index = ['T', 'F'])
```

```python
pd.DataFrame(confusion_matrix(y_test, cnb.predict(X_test)), columns = ['T', 'F'], index = ['T', 'F'])
```
