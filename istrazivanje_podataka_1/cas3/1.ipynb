```python
data.loc[data['SREDNJE_GOD_STANJE_EUR'] < 0, 'SREDNJE_GOD_STANJE_EUR'] = np.nan
# data
data.dropna(inplace = True)
# data
```

```python
data['STAROST'] = pd.cut(data['STAROST'], bins = 5, labels = [1, 2, 3, 4, 5])
```

```python
unique_values = data['BRACNO_STANJE'].unique()
unique_values
```

```python
data.loc[data['BRACNO_STANJE'] == 'samac', 'BRACNO_STANJE'] = 'nije_u_braku'
data.loc[data['BRACNO_STANJE'] == 'razveden(a)', 'BRACNO_STANJE'] = 'nije_u_braku'
```

```python
# standardizacija
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

num_atr = (data.dtypes[data.dtypes == 'float64'].index).append(data.dtypes[data.dtypes == 'int64'].index)
data[num_atr] = scaler.fit_transform(data[num_atr])
```
