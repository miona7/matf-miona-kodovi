```python
import os
```

```python
def read_data(root_dir):
    maps = []
    classes = []
    for class_name in os.listdir(root_dir):
        class_dir = os.path.join(root_dir, class_name)
        for file in os.listdir(class_dir):
            file_path = os.path.join(class_dir, file)
            word_counts = {}
            with open(file_path, 'r') as f:
                for line in f:
                    word, count = line.split()
                    word_counts[word] = int(count)
            maps.append(word_counts)
            classes.append(class_name)
    return maps, classes
```

```python
X_train, y_train = read_data('skupovi/Trening')
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer()
```

```python
dv.fit(X_train)
```

```python
sparse_matrix = dv.transform(X_train)
```

```python
import pandas as pd
```

```python
X_train = pd.DataFrame(sparse_matrix.toarray(), columns = dv.feature_names_)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB()
```

```python
mnb.fit(X_train, y_train)
```

```python
mnb.class_count_
```

```python
mnb.feature_count_
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
pd.DataFrame(confusion_matrix(y_train, mnb.predict(X_train)), columns = mnb.classes_, index = mnb.classes_)
```

```python
accuracy_score(y_train, mnb.predict(X_train))
```

```python
X_test, y_test = read_data('skupovi/Testing')
```

```python
X_test = pd.DataFrame(dv.transform(X_test).toarray(), columns = dv.feature_names_)
```

```python
pd.DataFrame(confusion_matrix(y_test, mnb.predict(X_test)), columns = mnb.classes_, index = mnb.classes_)
```

```python
accuracy_score(y_test, mnb.predict(X_test))
```
