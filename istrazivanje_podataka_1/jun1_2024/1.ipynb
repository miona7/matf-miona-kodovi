```python
import pandas as pd
```

```python
data = pd.read_csv('20newsgroups.csv')
```

```python
X = data.text
y = data.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.75, random_state = 123, stratify = y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
X_train = cv.fit_transform(X_train)
X_test = cv.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB()
```

```python
mnb.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
pd.DataFrame(confusion_matrix(y_train, mnb.predict(X_train)))
```

```python
accuracy_score(y_train, mnb.predict(X_train))
```

```python
accuracy_score(y_test, mnb.predict(X_test))
```

```python
pd.DataFrame(confusion_matrix(y_test, mnb.predict(X_test)))
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf = TfidfTransformer()
```

```python
tfidf.fit(X_train)
```

```python
X_train = tfidf.transform(X_train)
X_test = tfidf.transform(X_test)
```

```python
accuracy_score(y_test, mnb.predict(X_test))
```
