```python
def minimum_i_maximum(a, b):
    minimum = min(a, b)
    maksimum = max(a, b)
    return minimum, maksimum # (minimum, maksimum)

minimum, maksimum = minimum_i_maximum(5.5, 10) # (minimum, maksimum)
print(f"Minimum: {minimum}, Maksimum: {maksimum}")
```

```python
class Vozilo:
    def __init__(self, rb):
        print("kreiran je objekat klase Vozilo sa registarskim brojem " + rb)
        self.registarski_broj = rb 

    def __str__(self): 
        return f"Vozilo sa registarskim brojem {self.registarski_broj}"

    def vrati_registarski_broj(self):
        return self.registarski_broj
    
    def postavi_registarski_broj(self, rb): 
        self.registarski_broj = rb

    def vrati_tip_vozila(self):
        pass


# dozvoljeno visestruko nasledjivanje
# ukoliko klasa nasledjuje drugu klasu, nasledjena klasa se navodi u zagradi pri deklaraciji podklase
class Automobil(Vozilo):
    def __init__(self, rb, broj_tockova):
        print("kreiran je objekat klase Automobil sa registarskim brojem " + rb)
        # konstruktor nadklase se ne poziva automatski
        super().__init__(rb)
        self.broj_tockova = broj_tockova

    def vrati_tip_vozila(self):
        return "Automobil"
    
vozilo1 = Vozilo('BG-123-AB')
print(vozilo1.vrati_tip_vozila())

automobil1 = Automobil('BG-234-BC', 4)
print(automobil1.vrati_tip_vozila())
```

```python
class Vozilo:
    broj_vozila = 0

    def __init__(self, rb):
        print("kreiran je objekat klase Vozilo sa registarskim brojem " + rb)
        self.registarski_broj = rb 
        Vozilo.broj_vozila += 1

    def __str__(self): 
        return f"Vozilo sa registarskim brojem {self.registarski_broj}"

    def vrati_registarski_broj(self):
        return self.registarski_broj
    
    def postavi_registarski_broj(self, rb): 
        self.registarski_broj = rb

    def vrati_tip_vozila(self):
        pass


class Automobil(Vozilo):
    def __init__(self, rb, broj_tockova):
        print("kreiran je objekat klase Automobil sa registarskim brojem " + rb)
        super().__init__(rb)
        self.broj_tockova = broj_tockova

    def vrati_tip_vozila(self):
        return "Automobil"
    
vozilo1 = Vozilo('BG-123-AB')
automobil1 = Automobil('BG-234-BC', 4)

print(f"Broj vozila: {Vozilo.broj_vozila}")
```

```python
class Vozilo:
    broj_vozila = 0

    def __init__(self, rb):
        print("kreiran je objekat klase Vozilo sa registarskim brojem " + rb)
        self.registarski_broj = rb 
        Vozilo.broj_vozila += 1
        Vozilo.__update(self)

    def __str__(self): 
        return f"Vozilo sa registarskim brojem {self.registarski_broj}"

    def vrati_registarski_broj(self):
        return self.registarski_broj
    
    def postavi_registarski_broj(self, rb): 
        self.registarski_broj = rb

    def update(self):
        print("Vozilo.update")

    __update = update # __Vozilo__update - name mangling, metoda se moze pozvati preko self._Vozilo__update() ali ne i preko Vozilo.__update() jer je __update "privatna" metoda

    def vrati_tip_vozila(self):
        pass


class Automobil(Vozilo):
    def __init__(self, rb, broj_tockova):
        print("kreiran je objekat klase Automobil sa registarskim brojem " + rb)
        super().__init__(rb)
        self.broj_tockova = broj_tockova
        Automobil.__update(self)

    def update(self):
        print("Automobil.update")

    __update = update

    def vrati_tip_vozila(self):
        return "Automobil"
    
vozilo1 = Vozilo('BG-123-AB')
automobil1 = Automobil('BG-234-BC', 4)

print(f"Broj vozila: {Vozilo.broj_vozila}")
```
