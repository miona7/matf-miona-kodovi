```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 100 ekvidistantnih tacaka izmedju 1 i 5
x = np.linspace(1, 5, 100)

# funkcije
y0 = x
y1 = x * np.log(x)
y2 = np.sqrt(x)

plt.figure(figsize = (8, 4))
plt.plot(x, y0, label = 'y = x', linewidth = 2)
plt.plot(x, y1, label = 'y = x*log(x)', linewidth = 2)
plt.plot(x, y2, label = 'y = sqrt(x)', linewidth = 2)
plt.title('grafici funkcija')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha = 0.3)
plt.legend()
plt.show()
```

```python
                # pravi generator slucajnih brojeva sa semenom 7
rng = np.random.default_rng(7)
# 80 slucajnih brojeva iz normalne raspodele
x = rng.normal(0, 1, 80)
# zavisnost izmedju x i y sa dodatkom normalne buke (suma)
y = 0.7 * x + rng.normal(0, 0.6, 80)

plt.figure(figsize = (8, 4))
plt.scatter(x, y, alpha = 0.7, edgecolors = 'black')
plt.title('scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha = 0.3)
plt.show()
```

```python
                    # centar raspodele # standardna devijacija # broj uzoraka
samples = rng.normal(loc = 170, scale = 8, size = 300)

plt.figure(figsize = (8, 4))
plt.hist(samples, bins = 20, color = 'tab:blue',  edgecolor = 'black', alpha = 0.8)
plt.title('histogram nasumicnog uzorka')
plt.xlabel('vrednosti')
plt.ylabel('frekvencija')
# crta linije mreze samo na y-osi, sa prozirnoscu 0.3
plt.grid(axis = 'y', alpha = 0.3)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 200)

# ceo prozor tj platno  # mreza od 1 reda i 2 kolone
fig, axes = plt.subplots(1, 2, figsize = (10, 4))
    # pojedinacni grafici

axes[0].plot(x, np.sin(x), label = 'sin(x)', color = 'tab:green')
axes[0].set_title('sin(x)')
axes[0].grid(True, alpha = 0.3)

axes[1].plot(x, np.cos(x), label = 'cos(x)', color = 'tab:red')
axes[1].set_title('cos(x)')
axes[1].grid(True, alpha = 0.3)

fig.suptitle('sinus i kosinus')
plt.tight_layout()
plt.show()
```

```python
# nacrtati funkcije x^2 i x^3 na istom grafiku sa legendom
x = np.linspace(-5, 5, 200)

plt.figure(figsize = (8, 4))
plt.plot(x, x**2, label = 'y = x^2', color = 'tab:blue')
plt.plot(x, x**3, label = 'y = x^3', color = 'tab:orange')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha = 0.3)
plt.legend()
plt.show()
```

```python
# napraviti histogram nad 1000 uzoraka iz normalne raspodele

rng = np.random.default_rng(42)

samples = rng.normal(loc = 0, scale = 1, size = 1000)

plt.figure(figsize = (8, 4))
plt.hist(samples, bins = 30, color = 'tab:purple', edgecolor = 'black', alpha = 0.7, density = True)
plt.xlabel('vrednosti')
plt.ylabel('frekvencija')
plt.grid(axis = 'y', alpha = 0.3)
plt.show()
```

```python
# napraviti jedan scatter grafik sa bojom tacke koja zavisi od vrednosti x

rng = np.random.default_rng(123)
x = rng.uniform(-10, 10, 200)
y = 0.9 * x + rng.normal(0, 5, 200)

plt.figure(figsize = (8, 4))
plt.scatter(x, y, alpha = 0.7, edgecolors = 'black', c = x, cmap = 'plasma')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
