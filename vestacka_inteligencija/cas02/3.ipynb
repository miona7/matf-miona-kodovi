```python
species_categories = pd.Categorical(df['species'])

attributes = df.columns[:-1]
all_attributes_combinations = [attr_pair for attr_pair in combinations(attributes, 2)]
n = len(all_attributes_combinations) // 2

fig, ax = plt.subplots(n , 2, figsize = (10, 10))

for i, ax in enumerate(fig.axes):
    attr_comb = all_attributes_combinations[i]

    x = df[attr_comb[0]]
    y = df[attr_comb[1]]

    ax.scatter(x, y, c = species_categories.codes)
    ax.set_xlabel(attr_comb[0])
    ax.set_ylabel(attr_comb[1])

plt.tight_layout()
plt.show()
```
