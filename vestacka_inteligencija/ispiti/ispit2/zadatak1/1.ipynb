Na ispitu je data implementacija igre, a potrebno je implementirati MinMax sa AlphaBeta odsecanjem, funkciju end() i evaluate().

```python
import copy
import math

class NimState:
    def __init__(self, gomilice):
        self.gomilice = gomilice
        self.curr_player = 'X'

        self.last_move = None
        self.move_count = 0

    def play_move(self, move):
        i, uzmi = move[0], move[1]
        self.gomilice[i] -= uzmi
        self.curr_player = 'X' if self.curr_player == 'O' else 'O'
        self.last_move = move
        self.move_count += 1

    def draw_board(self):
        print(f"Gomilice: {self.gomilice} | Na potezu: {self.curr_player}")


def get_next_states(current_state: NimState):
    result = []
    for i, zetona in enumerate(current_state.gomilice):
        for uzmi in range(1, zetona + 1):
            next_state = copy.deepcopy(current_state)
            next_state.play_move([i, uzmi])
            result.append(next_state)
    return result

def get_winner(current_state: NimState):
    if sum(current_state.gomilice) == 0:
        return 'O' if current_state.curr_player == 'X' else 'X'
    return None

def evaluate(current_state: NimState):
  winner = get_winner(current_state)

  if winner == 'X':
    return 1000
  elif winner == 'O':
    return -1000

  return 0

def end(current_state: NimState):
  return sum(current_state.gomilice) == 0

def Min(current_state: NimState, alpha = -math.inf, beta = math.inf):
  if end(current_state):
    return evaluate(current_state), current_state

  best_move = None
  current_best_move_value = float("inf")

  for next_state in get_next_states(current_state):
    opponents_best_move_value, _ = Max(next_state, alpha, beta)

    if opponents_best_move_value < current_best_move_value:
      current_best_move_value = opponents_best_move_value
      best_move = next_state

    if current_best_move_value <= alpha:
      return current_best_move_value, next_state

    if current_best_move_value < beta:
      beta = current_best_move_value

  return current_best_move_value, best_move

def Max(current_state: NimState, alpha = -math.inf, beta = math.inf):
  if end(current_state):
    return evaluate(current_state), current_state

  best_move = None
  current_best_move_value = float("-inf")

  for next_state in get_next_states(current_state):
    opponents_best_move_value, _ = Min(next_state, alpha, beta)

    if opponents_best_move_value > current_best_move_value:
      current_best_move_value = opponents_best_move_value
      best_move = next_state

    if current_best_move_value >= beta:
      return current_best_move_value, next_state

    if current_best_move_value > alpha:
      alpha = current_best_move_value

  return current_best_move_value, best_move

def read_next_move_from_stdin():
    move = input("Unesi potez (indeks_gomilice, broj_zetona): ").split(',')
    return [int(move[0]), int(move[1])]

def get_next_computer_move(current_state: NimState, func):
    _, state = func(current_state)
    return state.last_move


if __name__ == '__main__':
    game = NimState([2, 3, 1])
    game.draw_board()

    while True:
        next_move = read_next_move_from_stdin()
        game.play_move(next_move)
        game.draw_board()

        if get_winner(game) == 'X':
            print('Player X won!')
            break
        if end(game):
            print('Tie!')
            break

        number_of_calls = 0
        next_move = get_next_computer_move(game, Min)
        print('Kompjuter igra potez:', next_move)
        game.play_move(next_move)
        game.draw_board()

        if get_winner(game) == 'O':
            print('Player O won!')
            break
```
