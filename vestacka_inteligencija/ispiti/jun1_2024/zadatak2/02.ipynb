```python
from z3 import *
# DeclareSort, BoolSort, Function, ForAll, Exists, Solver, Const, Consts, And, Or, Implies
```

```python
Person = DeclareSort('Person')

Invited = Function('Invited', Person, BoolSort()) # osoba x je pozvana
InPhonebook = Function('InPhonebook', Person, Person, BoolSort()) # osoba x u svom imeniku ima osobu y
Friend = Function('Friend', Person, BoolSort()) # osoba x je Nikolin prijatelj
```

```python
nikola, ivan, stefan, nemanja, dragan, andrija = Consts('Nikola Ivan Stefan Nemanja Dragan Andrija', Person)
x, y = Consts('x y', Person)
```

```python
s = Solver()

axioms = [
    Distinct(nikola, ivan, stefan, nemanja, dragan, andrija),

    Friend(ivan),
    Friend(stefan),
    Friend(nemanja),
    Friend(dragan),
    Friend(andrija),

    InPhonebook(nikola, ivan),
    InPhonebook(stefan, ivan),
    InPhonebook(nemanja, dragan),
    InPhonebook(nemanja, ivan),
    InPhonebook(nemanja, nikola),
    InPhonebook(dragan, nemanja),
    InPhonebook(dragan, andrija),
    InPhonebook(ivan, nemanja),
    InPhonebook(ivan, stefan),
    InPhonebook(andrija, nikola),

    Invited(ivan),
    ForAll([x, y], Implies(And(Invited(x), InPhonebook(x, y), Friend(y)), Invited(y)))
    # za svako x, y (Invited(x) && InPhonebook(x, y) && Friend(y)) => Invited(y)
]

s.add(axioms)
```

```python
if s.check() == sat:
    conjecture = And(Invited(ivan), Invited(stefan), Invited(nemanja), Invited(dragan), Invited(andrija))
    s.add(Not(conjecture))
    
    result = s.check()
    print(result)

    if result == unsat:
        print('conjecture proved')
    else:
        print('unsat')
```
