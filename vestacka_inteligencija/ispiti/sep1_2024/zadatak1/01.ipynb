```python
import math
from collections import defaultdict
```

```python
def load(filename):
    with open(filename, 'r') as f:
        return [line.strip() for line in f.readlines()]

def print_teren(t):
    for red in t:
        print(red)

def find(teren, name):
    n = len(teren)
    m = len(teren[0])
    for i in range(n):
        for j in range(m):
            if teren[i][j] == name:
                return (i, j)
```

```python
def euclidean(a, b):
    return math.sqrt((a[0] - b[0])**2 + (a[1] - b[1])**2)

def manhattan(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def chebyshev(a, b):
    return max(abs(a[0] - b[0]), abs(a[1] - b[1]))
```

```python
def get_neighbours(u, n, m, teren):
    i, j = u[0], u[1]
    moves = [(0, 1), (0, -1), (1, 0), (-1, 0)]
    neighbours = []

    for move in moves:
        di, dj = i + move[0], j + move[1]

        if 0 <= di < n and 0 <= dj < m and teren[di][dj] != '#':
            neighbours.append((di, dj))

    return neighbours
```

```python
def astar(teren, start, target, h):
    iterations = 0
    path = []
    n = len(teren)
    m = len(teren[0])
    
    open_list = set([start])
    closed_list = set()

    g = defaultdict(lambda: float("inf"))
    g[start] = 0

    parent = {start: None}

    while open_list:
        u = min(open_list, key = lambda v: g[v] + h(v, target))

        if u is None:
            break

        iterations += 1

        if u == target:

            while u is not None:
                path.append(u)
                u = parent[u]

            path.reverse()
            return iterations, path

        for v in get_neighbours(u, n, m, teren):
            if v not in open_list and v not in closed_list:
                open_list.add(v)
                parent[v] = u
                g[v] = g[u] + 1
            else:
                if g[v] > g[u] + 1:
                    parent[v] = u
                    g[v] = g[u] + 1

                    if v in closed_list:
                        closed_list.remove(v)
                        open_list.add(v)

        open_list.remove(u)
        closed_list.add(u)
    
    return iterations, path
```

```python
if __name__ == '__main__':
    teren = load('teren.txt')
    # print_teren(teren)

    A_coords = find(teren, 'A')
    B_coords = find(teren, 'B')
    # print(A_coords, B_coords)
    
    results = []
    for name, heuristic in [('Euklidska', euclidean), ('Menhetn', manhattan), ('Cebisevljeva', chebyshev)]:
        iterations, path = astar(teren, A_coords, B_coords, heuristic)
        results.append((iterations, name, path))
        print(f'{name}: iteracije = {iterations}, duzina puta = {len(path)}')

    best_iterations, best_name, best_path = min(results, key = lambda item: item[0])
    print('\nNajbolja heuristika:', best_name)
    print('Broj iteracija:', best_iterations)
    print('Put:', best_path)
```
