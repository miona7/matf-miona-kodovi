```python
Node = DeclareSort('Node')

Neighbour = Function('Neighbour', Node, Node, BoolSort()) # cvor x je sused cvoru y
Path = Function('Path', Node, Node, BoolSort()) # od cvora x do cvora y postoji put
```

```python
a, b, c, d, e, f, g, h, m, k = Consts('a b c d e f g h m k', Node)
x, y, z = Consts('x y z', Node) 
```

```python
axioms = [
    Distinct(a, b, c, d, e, f, g, h, m, k),

    Neighbour(a, b),
    Neighbour(b, c),
    Neighbour(b, d),
    Neighbour(d, g),
    Neighbour(d, e),
    Neighbour(e, f),
    Neighbour(h, m),
    Neighbour(m, k),

    # postoji put od čvora X do čvora Y ako je X sused od Y ili ako postoji sused Z od Y, takav da postoji put od X do Z
    ForAll([x, y], Implies(Neighbour(x, y), Path(x, y))),
    ForAll([x, y, z], Implies(And(Path(x, z), Neighbour(z, y)), Path(x, y))),

    # ako je X sused od Y, onda je i Y sused od X
    ForAll([x, y], Implies(Neighbour(x, y), Neighbour(y, x))),

    # ako postoji put od X do Y, onda postoji i put od Y do X
    ForAll([x, y], Implies(Path(x, y), Path(y, x)))
]
```

```python
s = Solver()
s.add(axioms)
```

```python
if s.check() == sat:
    conjecture = Path(a, f)

    s.add(Not(conjecture))

    result = s.check()

    if result == unsat:
        print('conjecture proved')
    else:
        print(s.model())
else:
    print('contradictor axioms')        
```
